```python
import re


def query_to_regex(input_query):
    input_query = input_query.replace('"', '')
    input_query = input_query.replace(' OR ', '|')
    parts = re.split(r' AND ', input_query)
    modified_parts = [rf'(?=.*\b{part.strip()}\b)' for part in parts]
    regex_query = ''.join(modified_parts)
    return regex_query

df_query['regex_query'] = df_query['keyword'].apply(query_to_regex)
combined_regex = re.compile('|'.join(df_query['regex_query'].astype(str)))
```

```python
# ("инновационная деятельность") AND ("фактор" OR "причины успеха" OR "ввиду" OR "благодаря" OR "из-за того что" OR "привело к тому что") AND ("компания" OR "фирма" OR "предприятие" OR "корпорация")

text = 'бизнес-инкубатор изобреение фирма нововведение кккау фактор'

if combined_regex.search(text):
    print("Совпадение найдено!")
else:
    print("Совпадение не найдено.")
```

```python
regex_queries = []

for index, row in df_query.iterrows():
    regex_queries.append(query_to_regex(row['keyword']))

df_query['regex_query'] = regex_queries
combined_regex = re.compile('|'.join(df_query['regex_query'].astype(str)))
```
